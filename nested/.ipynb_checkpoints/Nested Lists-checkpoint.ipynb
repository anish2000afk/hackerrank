For making a list of records 

```python
records = []
for _ in range(int(input())):
    name = input()
    score = float(input())
    records.append([name, score])
print(sorted((records)))
```

To find if we can sort a list of floats 

```python
floats = [1.0, 2.0, 3.0, 2.5, 201.0]
print(sorted(floats))
```

```python
''' This happens because the list when populating has only item inside on the first instance'''
''' The above problem can be solved by moving second last outside the list '''
```

```python
records = []
u =[]
score_list = []
for _ in range(int(input())):
    name = input()
    score = float(input())
    records.append([name, score])
    score_list.append(score)
    u = set(list(u)).union(set(score_list))
    print(u) 
second_last = list(u)[-2] # moved the list outside the for loop.
print(second_last)
```

The problem can also be solved by making a if else statement with len()

```python
records = []
u =[]
score_list = []
for _ in range(int(input())):
    name = input()
    score = float(input())
    records.append([name, score])
    score_list.append(score)
    u = set(list(u)).union(set(score_list))
    print(u)
    if len(u) == 1:
        second_last = list(u)[-1] # Convert set to list and grab the -2 index.
    else:
        second_last = list(u)[-2]
    # names = [i for j in records for i in j if i == second_last]
print(second_last)
```
